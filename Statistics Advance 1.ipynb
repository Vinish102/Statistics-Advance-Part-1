```python
"""
A random variable in probability theory is a variable that can take on different values, 
each with an associated probability. 
It is a function that assigns a numerical value to each outcome in a sample space. 
Random variables can be discrete, taking on a finite or countable number of values, 
or continuous, taking on an infinite number of values within a range.
"""
```

```python
#2  What are the types of random variable?
```

```python
"""
There are two main types of random variables:

    Discrete Random Variables: These take on a finite or countable number of distinct values. 
    Examples include the number of heads in a series of coin flips or the roll of a die.

    Continuous Random Variables: These can take on an infinite number of values within a given range. 
    Examples include the exact height of individuals or the time it takes to run a race.


"""
```

```python
#3  What is the difference between discrete and continuous distributions?
```

```python
"""
The difference between discrete and continuous distributions lies in the type of random variables they describe:

    Discrete Distributions: These are associated with discrete random variables. 
    They describe the probability of each possible value that the variable can take. 
    Examples include the binomial distribution and the Poisson distribution.

    Continuous Distributions: These are associated with continuous random variables. 
    They describe the probability of the variable falling within a certain range of values, 
    often represented by a probability density function (PDF). 
    Examples include the normal distribution and the exponential distribution.


"""
```

```python
#4 What are probability distribution functions (PDF)?
```

```python
"""
Probability Distribution Functions (PDF) describe the likelihood of a random variable taking on a particular value or range of values. 
There are two types:

    Probability Mass Function (PMF): Used for discrete random variables, it gives the probability that a discrete random variable is exactly equal to some value.

    Probability Density Function (PDF): Used for continuous random variables, it describes the relative likelihood of the variable taking on a value within a continuous range. 
    The area under the curve of a PDF over an interval represents the probability of the variable falling within that interval.


"""
```

```python
#5  How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
```

```python
"""
Cumulative Distribution Functions (CDF) and Probability Distribution Functions (PDF) serve different purposes:

    Probability Distribution Function (PDF):
        For discrete variables, it’s called a Probability Mass Function (PMF).
        For continuous variables, it describes the likelihood of the variable taking on a specific value or range.
        The PDF does not give probabilities directly for continuous variables; instead, the area under the curve over an interval gives the probability.

    Cumulative Distribution Function (CDF):
        Applies to both discrete and continuous variables.
        Represents the probability that a random variable is less than or equal to a certain value.
        It is a non-decreasing function that ranges from 0 to 1.

The CDF provides cumulative probabilities, while the PDF provides density or mass at specific points or intervals.

"""
```

```python
 #13 What is the Central Limit Theorem (CLT), and why is it critical in statistics?
```

```python
"""
The Central Limit Theorem (CLT) is a fundamental principle in statistics that states that the distribution of the sample mean 
of a sufficiently large number of independent, 
identically distributed random variables approaches a normal distribution, regardless of the original distribution of the variables.

Key reasons for its critical importance include:

    Normal Approximation: It allows for the approximation of the sampling distribution of the mean to be normal, facilitating statistical inference.

    Applicability: It applies to various types of data and distributions, making it versatile in real-world applications.

    Basis for Inference: It underpins many statistical methods, including hypothesis testing and confidence intervals, by enabling the use of normal distribution properties.


"""
```

```python
#14 How does the Central Limit Theorem relate to the normal distribution?
```

```python
"""
The Central Limit Theorem (CLT) relates to the normal distribution by stating that the distribution of the sample mean
of a large number of independent, identically distributed random variables will approximate a normal distribution, 
regardless of the original distribution of the variables.

This means that even if the underlying data is not normally distributed, 
the distribution of the sample means will tend to be normal as the sample size increases. 
This property allows statisticians to use normal distribution techniques for inference, 
such as calculating confidence intervals and conducting hypothesis tests, 
even when the data itself is not normally distributed.

"""
```

```python
#15  What is the application of Z statistics in hypothesis testing?
```

```python
"""
In hypothesis testing, Z statistics are used to determine whether to reject a null hypothesis by comparing the observed data 
to what is expected under the null hypothesis. The key applications include:

    Standardization: Z statistics convert data to a standard normal distribution, 
    allowing comparison across different datasets.

    Z-Test: Used for testing hypotheses about population means when the population standard deviation is known and the sample size is large. 
    It calculates the Z-score, which measures how many standard deviations an element is from the mean.

    P-Value Calculation: The Z-score helps determine the p-value, which indicates the probability of observing the data if the null hypothesis is true.
    A low p-value suggests rejecting the null hypothesis.


"""
```

```python
#16 How do you calculate a Z-score, and what does it represent?
```

```python
"""
The Z-score represents the number of standard deviations a data point is from the mean. 
A positive Z-score indicates the data point is above the mean, while a negative Z-score indicates it is below the mean. 
Z-scores allow for comparison between different data points and help identify how unusual or typical a data point is within a distribution.
A Z-score is calculated using the formula:

Z=(X−μ)/σ

where:

    ( X ) is the value of the data point,
    ( \mu ) is the mean of the population,
    ( \sigma ) is the standard deviation of the population.

"""
```

```python
"""
In statistics, point estimates and interval estimates are used to infer population parameters from sample data:

    Point Estimates: These provide a single value as an estimate of a population parameter. 
    For example, the sample mean is a point estimate of the population mean.

    Interval Estimates: These provide a range of values within which the population parameter is expected to lie, 
    with a certain level of confidence. Confidence intervals are a common type of interval estimate, 
    indicating the reliability of the estimate by specifying a range and a confidence level (e.g., 95%).


"""
```

```python
"""
The Central Limit Theorem (CLT) relies on certain assumptions to hold true:

    Independence: The random variables must be independent of each other.

    Identically Distributed: The random variables should be identically distributed, meaning they have the same probability distribution.

    Sample Size: The sample size should be sufficiently large. While there’s no strict rule, a common guideline is a sample size of 30 or more.

    Finite Variance: The random variables should have a finite variance.

When these conditions are met, the CLT ensures that the distribution of the sample mean approaches a normal distribution as the sample size increases.

"""
```

```python
#3 Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
```

```python
def bernoulli_pdf(x, p):
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    return p**x * (1 - p)**(1 - x)

# Example usage:
p = 0.5  # Probability of success
x = 1    # Outcome (0 or 1)
probability = bernoulli_pdf(x, p)
print("Probability:", probability)
```

```python
#19 Implement hypothesis testing using Z-statistics for a sample dataset
```

```python

# Sample dataset
data = np.random.normal(50, 10, 30)  # Sample data with mean=50, std=10

# Population parameters
population_mean = 52  # Hypothesized population mean
population_std = 10   # Known population standard deviation

# Calculate sample mean and size
sample_mean = np.mean(data)
sample_size = len(data)

# Calculate Z-statistic
z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Determine p-value for a two-tailed test
p_value = 2 * (1 - norm.cdf(abs(z_statistic)))

# Significance level
alpha = 0.05

# Decision
if p_value < alpha:
    print(f"Reject the null hypothesis (p-value: {p_value:.4f})")
else:
    print(f"Fail to reject the null hypothesis (p-value: {p_value:.4f})")

print(f"Z-statistic: {z_statistic:.4f}")
```

```python
#20  Create a confidence interval for a dataset using Python and interpret the result
```

```python
import numpy as np
from scipy.stats import norm

# Sample dataset
data = np.random.normal(50, 10, 100)  # Sample data with mean=50, std=10

# Calculate sample mean and standard deviation
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Use ddof=1 for sample standard deviation
sample_size = len(data)

# Confidence level
confidence_level = 0.95
alpha = 1 - confidence_level

# Critical value for a 95% confidence interval
z_critical = norm.ppf(1 - alpha/2)

# Margin of error
margin_of_error = z_critical * (sample_std / np.sqrt(sample_size))

# Confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: {confidence_interval[0]:.2f} to {confidence_interval[1]:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate two datasets
data1 = np.random.normal(50, 10, 1000)  # Dataset 1 with mean=50, std=10
data2 = np.random.normal(50, 15, 1000)  # Dataset 2 with mean=50, std=15

# Calculate standard deviations
std_dev1 = np.std(data1, ddof=1)
std_dev2 = np.std(data2, ddof=1)

# Print standard deviations
print(f"Standard Deviation of Dataset 1: {std_dev1:.2f}")
print(f"Standard Deviation of Dataset 2: {std_dev2:.2f}")

# Visualize the difference using a bar plot
labels = ['Dataset 1', 'Dataset 2']
std_devs = [std_dev1, std_dev2]

plt.bar(labels, std_devs, color=['blue', 'orange'])
plt.title('Comparison of Standard Deviations')
plt.ylabel('Standard Deviation')
plt.ylim(0, max(std_devs) + 5)
plt.show()
```

```python
#26  Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution
```

```python
import numpy as np

# Generate data from a normal distribution
mean = 50
std_dev = 10
sample_size = 1000
data = np.random.normal(mean, std_dev, sample_size)

# Calculate the range
data_range = np.max(data) - np.min(data)

# Calculate the interquartile range (IQR)
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

print(f"Range: {data_range:.2f}")
print(f"Interquartile Range (IQR): {iqr:.2f}")
```

```python
#27  Implement Z-score normalization on a dataset and visualize its transformation
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data from a normal distribution
mean = 50
std_dev = 10
sample_size = 1000
data = np.random.normal(mean, std_dev, sample_size)

# Z-score normalization
def z_score_normalization(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    return (data - mean) / std_dev

normalized_data = z_score_normalization(data)

# Visualize the original and normalized datasets
plt.figure(figsize=(12, 5))

# Original data
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Original Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Normalized data
plt.subplot(1, 2, 2)
plt.hist(normalized_data, bins=30, color='salmon', edgecolor='black', alpha=0.7)
plt.title('Z-score Normalized Data')
plt.xlabel('Z-score')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
#28 Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution
```

```python
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(data):
    data_skewness = skew(data)
    data_kurtosis = kurtosis(data)
    return data_skewness, data_kurtosis

# Generate data from a normal distribution
mean = 50
std_dev = 10
sample_size = 1000
data = np.random.normal(mean, std_dev, sample_size)

# Calculate skewness and kurtosis
skewness, kurtosis_value = calculate_skewness_kurtosis(data)

print(f"Skewness: {skewness:.4f}")
print(f"Kurtosis: {kurtosis_value:.4f}")
```
